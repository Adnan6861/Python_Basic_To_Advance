# 1. Why Do We Need Conditional Statements?

Conditional statements are used to make decisions in a program based on specific conditions. They allow the program to execute different blocks of code depending on whether a condition is true or false.

# Real-Life Examples:

If it rains, take an umbrella.
If the traffic light is green, drive; if it is red, stop.
If the user enters the correct password, allow login; otherwise, deny access.
Key Insight: Conditional statements change the sequence of program execution based on conditions.

# condition also change the Sequence of Programm that are executed 

```python
#Code
name="Adnan"
if name =="Adnan":  # condition checks
    print('Its Adnan')
else:
    print("Unkown")
```

```python
#use_case of elif
#if i have multiple condation to Execute so we used elif
day="sunday"
if day=='Tue':
    print("Not Your Day")
elif day == 'sunday':
    print('YOur Day Buddy')
else:                     # else statement are Execute when all the Statment are False
    print('Its Working Day')
```

```python
# Neste Conditions --> condition under condations
name='Adnan'
roll=189
degree='IT'
if name == 'Adnan':
    if degree=='IT':
        if roll==189:                #if all nested loop statisfied than if state return value other wise else Execute
            print('this is my student')
        else:
            print('Outsider')
```

```python
# logical Operator  and ,or
animal = 'Dog'
age=7
if animal == 'Dog' and  age == 9: # if both are true than if is excuted other wise else are Executed
    print("its My Dog")
else:
    print("i don't know")
```
