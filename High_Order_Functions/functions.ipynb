# High_Order_Function
 the Function take a input other function also return as function as a output

```python
# map()
def double(x):
    return x*2

number=[10,20,30,40,50,60]
values=list(map(double,number))
print(values)

# lamda function
number=[10,20,30,40,50]
type_convertor=list(map(lambda x:x/2,number))
print(type_convertor)
```

```python
#  fliter

def even(x):
    if x%2==0:
        return x
number_doer=list(filter(even,number))
print(number)
```

```python
from functools import reduce
def multple(x,y):
    return x*y
f_call=reduce(multple,number)
print(f_call)
```
