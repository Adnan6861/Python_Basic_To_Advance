 # Loops in Python

Loops are used to perform repetitive tasks efficiently. Python provides two types of loops:

For Loop: Iterates over a sequence (e.g., list, tuple, dictionary).
While Loop: Executes a block of code as long as a condition is true.
For Loop Syntax

for item in sequence:

    # Code to execute

```python
# For Loop
number =[1,2,36,7,8]   # list of numbers
                       # indexing system are started from 0 
for i in number:
    print(i)      # i goes each element in number-list also Print Number one by one
             
    
    
```

# Enumerate() in For Loops:

```python
#Manualy Enumerate 
fruit_list=["Apple","Mango","Banana","Cherry"]
index=0                                           #same thing also Do with enumerate_Function
for i in fruit_list:
    print(f"{index} : {i}")
    index+=1
```

```python
#code
list_of_somethings=['Adnan','Imran','Kamran']
for index,value in enumerate(list_of_somethings):  # enumerate are also print the index with elements 
    print(f'{index} : {value}')
```

# While Loop

Specific Work repete multiple time on the basis of true and Flase

while loop run as specfied condation is true

eg: pasword in login when we enter incorrect password they again and again asked password when password is true they stop and execute main tak

```python
#code
count=0
while count < 4 :
    print(f"value:{count}") # when value are true the loop are excuted . when Values is False it is terminates
    # take care in while Loop the loop are not goes to infinate it values always be true
    count+=1
```

# Loop Control statement
  we also Control the Loop by specific statement or condations 
  for example break

```python
# Continue Statement : you Skip specifc ilteratio and Move to Next element it also Used Continue Statement
# we skip one or more element but loop are not stop
for i in range(1,11):
    if i==3:               # it also Skip the 3 element 
        continue
    print(i)
else:
    print('Loop is executed and skip the element 3')
    
#else condation used at the end of code  it execute when naturally end loop
```

```python
# break are used to exit the loop at specific element 
count=0
while count < 10:
    if count == 9:
        break
    count+=1
else:
    print("Code is Executed Fully")
```

```python
# Saqure 
for i in range(6):
    for j in range(6):
        print('*',end=" ")
    print()
```

```python
#right angle
for i in range(5):
    for j in range(i+1):
        print(j,end=" ")
    print()
     
```

```python
#inverted
for i in range(1,10):
    for j in range(10-i):
        print("*",end=" ")
    print()
```

```python
#left angle
for rows in range(5):
    for i in range(5-rows-1):
        print(' ',end="")
    for k in range(rows+1):
        print('*',end="")
    print()
    
```

```python
ele=10
for rows in range(ele):
    for space in range(rows):
        print(" ",end="")
    for star in range(ele-rows):
        print("*",end="")
    print()
```

```python
# upper Pyramid
def upper_rows(rows):
    for i in range(rows):
        for j in range(rows-i-1):
            print(" ",end="")
        for k in range(2*i+1):
            print("*",end="")
        print()
print(upper_rows(5))
```
