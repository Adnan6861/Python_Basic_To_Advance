 # Loops in Python

Loops are used to perform repetitive tasks efficiently. Python provides two types of loops:

For Loop: Iterates over a sequence (e.g., list, tuple, dictionary).
While Loop: Executes a block of code as long as a condition is true.
For Loop Syntax

for item in sequence:

    # Code to execute

```python
# For Loop
number =[1,2,36,7,8]   # list of numbers
                       # indexing system are started from 0 
for i in number:
    print(i)      # i goes each element in number-list also Print Number one by one
             
    
    
```

# Enumerate() in For Loops:

```python
#Manualy Enumerate 
fruit_list=["Apple","Mango","Banana","Cherry"]
index=0                                           #same thing also Do with enumerate_Function
for i in fruit_list:
    print(f"{index} : {i}")
    index+=1
```

```python
#code
list_of_somethings=['Adnan','Imran','Kamran']
for index,value in enumerate(list_of_somethings):  # enumerate are also print the index with elements 
    print(f'{index} : {value}')
```

# For Loop iltrate on Dictionary:

```python
student={
    'name':'Adnan',
    'roll_number':189,
    'email':'adnannawaxk080@gmail.com',
    'age':19
    
}
for key in student:  # only print the Keys in dictionary
    print(key)


    
```

```python
#print both Key and values
for key in student: 
    
    print(f'{key}:{student[key]}')
print('-'*30) 
# Second Method:
for key,value in student.items():
   print(f'{key}:{value}')
    
```
